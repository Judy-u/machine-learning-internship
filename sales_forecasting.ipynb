# Sales Forecasting System

## Objective
This project predicts future sales based on historical data using machine learning.

```python
import pandas as pd

data = {
    "month": list(range(1, 13)),
    "sales": [200, 220, 250, 270, 300, 320, 310, 305, 330, 360, 390, 420]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df["month"], df["sales"], marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show() 
```

```python
from sklearn.linear_model import LinearRegression

X = df[["month"]]
y = df["sales"]

model = LinearRegression()
model.fit(X, y)
```

```python
future_months = pd.DataFrame({
    "month": [13, 14, 15, 16, 17, 18]
})

predictions = model.predict(future_months)

future_months["predicted_sales"] = predictions
future_months
```

```python
plt.plot(df["month"], df["sales"], label="Past Sales")
plt.plot(future_months["month"], future_months["predicted_sales"], label="Predicted Sales", linestyle='dashed')

plt.title("Sales Forecasting")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
