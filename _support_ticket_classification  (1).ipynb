```python
import pandas as pd
```

```python
data = {
    "text": [
        "I forgot my password",
        "Unable to login to my account",
        "Please help me reset my password",
        "Payment failed during checkout",
        "I was charged twice",
        "Refund has not been received",
        "The app keeps crashing",
        "Website is not loading",
        "There is a bug in the dashboard",
        "My order has not arrived",
        "Delivery is delayed",
        "I received the wrong item"
    ],
    "category": [
        "Account",
        "Account",
        "Account",
        "Billing",
        "Billing",
        "Billing",
        "Technical",
        "Technical",
        "Technical",
        "Delivery",
        "Delivery",
        "Delivery"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split

X = df["text"]
y = df["category"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("Training data:")
print(X_train)

print("\nTesting data:")
print(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)

print(X_train_vectors.shape)
print(X_test_vectors.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_vectors, y_train)
```

```python
y_pred = model.predict(X_test_vectors)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
new_tickets = [
    "My payment did not go through",
    "I cannot sign into my account",
    "The mobile app crashes",
    "My package has not been delivered"
]

new_vectors = vectorizer.transform(new_tickets)
predictions = model.predict(new_vectors)

for ticket, pred in zip(new_tickets, predictions):
    print("Ticket:", ticket)
    print("Predicted category:", pred)
    print()
```
