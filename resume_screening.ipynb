# Resume Screening System

## Objective
This project ranks resumes based on how similar they are to a job description using machine learning.

```python
import pandas as pd

resumes = [
    "Python developer with experience in machine learning and data analysis",
    "Web developer skilled in HTML, CSS, JavaScript",
    "Data scientist with Python, pandas, and machine learning experience",
    "Mobile app developer using Flutter and Dart"
]

job_description = "Looking for a data scientist with Python and machine learning skills"
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

all_text = resumes + [job_description]

vectors = vectorizer.fit_transform(all_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

resume_vectors = vectors[:-1]
job_vector = vectors[-1]

similarity = cosine_similarity(resume_vectors, job_vector.reshape(1, -1))
```

```python
scores = similarity.flatten()

results = pd.DataFrame({
    "Resume": resumes,
    "Score": scores
})

results = results.sort_values(by="Score", ascending=False)
print(results.to_string())
```
